```python
from sklearn.datasets import make_regression
import numpy as np

# n_informative --> no. of relevant features which affects the o/p
x, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)

import matplotlib.pyplot as plt
plt.scatter(x, y)
```

```python
# THIS IS THE ANSWER OF COEFF AND INTERCEPT THAT WE SHOULD GET AFTER APPLYING GRADIENT DESCENT

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
reg = LinearRegression()

reg.fit(x, y)

plt.scatter(x, y)
plt.plot(x, reg.predict(x), color="red")

print(reg.coef_)
print(reg.intercept_)

```

```python
# APPLYING GRADIENT DESCENT ASSUMING SLOPE IS CONSTANT (78.35) AND STARTING VALUE OF C=0
y_pred = ((78.35 * x) + 0).reshape(4)

plt.scatter(x, y)
plt.plot(x, reg.predict(x), color="red")
plt.plot(x, y_pred, color="blue")

```

```python
# ITERATION - 01    
c = 0
m = 78.35

loss_slope = -2 * np.sum(y - m*x.ravel() - c)
# loss_slope

# Let learning rate be 0.2
lr = 0.1
step_size = lr*loss_slope
# step_size

c = c - step_size
# c

y_pred = ((78.35 * x) + c).reshape(4)

plt.scatter(x, y)
plt.plot(x, reg.predict(x), color="red")
plt.plot(x, y_pred, color="blue")
```

```python
# ITERATION - 02   

loss_slope = -2 * np.sum(y - m*x.ravel() - c)
# loss_slope

# Let learning rate be 0.2
lr = 0.1
step_size = lr*loss_slope
# step_size

c = c - step_size
# print(c)

y_pred = ((78.35 * x) + c).reshape(4)

plt.scatter(x, y)
plt.plot(x, reg.predict(x), color="red")
plt.plot(x, y_pred, color="blue")
```
