```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

x, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=40, hypercube=False, class_sep=10)


plt.figure(figsize=(10, 6))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='winter', s=100)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(x, y)

# y = Ax + By + C
# slope = m = -a/b
# intercept = c = -c/b

m = -(lor.coef_[0][0]/lor.coef_[0][1])
c = -(lor.intercept_/lor.coef_[0][1])

x_input = np.linspace(-3,3,100)
y_input = m*x_input + c

plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='black',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
