```python

from sklearn.datasets import fetch_california_housing
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR # support vector machines
from sklearn.model_selection import cross_val_score


lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()

x, y = fetch_california_housing(return_X_y=True)
estimators = [('lr',lr),('dt',dt),('svr',svr)]


for estimator in estimators:
  r = cross_val_score(estimator[1],x,y,scoring='r2',cv=10)
  print(estimator[0],np.round(np.mean(r),2))


from sklearn.ensemble import VotingRegressor
estimators = [('lr',lr),('dt',dt),('svr',svr)]

vr = VotingRegressor(estimators)
r = cross_val_score(vr,x,y,scoring='r2',cv=10)
print("Voting Regressor",np.round(np.mean(r),2))

for i in range(1,4):
  for j in range(1,4):
    for k in range(1,4):
      vr = VotingRegressor(estimators,weights=[i,j,k])
      r = cross_val_score(vr,x,y,scoring='r2',cv=10)
      print("For i={},j={},k={}".format(i,j,k),np.round(np.mean(r),2))

```
